# Práctica de Regresión Lineal con Pipeline

En este notebook se desarrolla una práctica de regresión lineal aplicada a un caso
de la vida real, utilizando un Pipeline de Scikit-learn.

El objetivo es analizar la relación entre el tamaño de un departamento y su precio,
siguiendo un flujo correcto de aprendizaje automático que incluye:
- Carga de datos desde un archivo CSV
- Escalado de variables
- Entrenamiento de un modelo de regresión lineal mediante un pipeline
- Visualización de resultados


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

## Carga del conjunto de datos

El conjunto de datos representa un caso realista donde se analiza cómo el tamaño
de un departamento influye directamente en su precio, mostrando un comportamiento
lineal.


```python
df = pd.read_csv("housing_linear_data.csv")
df.head()
```

## Definición de variables

- Variable independiente (X): Tamaño del departamento en metros cuadrados.
- Variable dependiente (y): Precio del departamento en miles de pesos mexicanos.


```python
X = df[['ApartmentSize_m2']]
y = df['Price_ThousandsMXN']
```

## Construcción del Pipeline

Se construye un pipeline que integra el preprocesamiento de los datos y el modelo
de regresión lineal en un solo flujo de trabajo.

El pipeline está compuesto por:
1. StandardScaler para el escalado de los datos.
2. LinearRegression como modelo predictivo.


```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('linear_regression', LinearRegression())
])
```

```python
pipeline.fit(X, y)
y_pred = pipeline.predict(X)
```

## Visualización de resultados

A continuación se muestran los valores reales del conjunto de datos junto con
la recta de regresión obtenida a partir del pipeline.


```python
plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred, color="red")
plt.xlabel("Tamaño del departamento (m²)")
plt.ylabel("Precio (miles de MXN)")
plt.title("Regresión Lineal usando Pipeline")
plt.show()
```

## Conclusiones

El uso de un Pipeline permite integrar de manera ordenada el escalado de los datos
y el entrenamiento del modelo de regresión lineal, reduciendo errores en el proceso.

Los resultados muestran una relación lineal clara entre el tamaño del departamento
y su precio, lo cual valida el uso de la regresión lineal para este tipo de problemas
de predicción.
