#Medición de tiempo con lista en Python

Este bloque crea una lista con un millón de números consecutivos usando range(). Posteriormente, se inicia un contador de tiempo para medir el rendimiento del ciclo for. El bucle recorre todos los elementos sin realizar ninguna operación, permitiendo calcular el tiempo que tarda Python en recorrer grandes volúmenes de datos.

```python
datos = list (range(1000000))
inicio = time.perf_counter()
for i in datos:
 pass
final = time.perf_counter()
tiempo = final - inicio
print(f"tiempo de ejecucion: {tiempo:.6f}sec")
```

#Rendimiento de operaciones con NumPy

Este bloque genera un arreglo numérico con la función arange() de NumPy y calcula la suma de sus elementos usando np.sum(). Además, se mide el tiempo de ejecución para analizar el rendimiento de NumPy al trabajar con grandes cantidades de datos.

```python
np_lista = np.arange(1000000000)

inicio = time.perf_counter()
suma_np = np.sum(np_lista)
final = time.perf_counter()

tiempo = final - inicio
print(f"Tiempo de ejecución: {tiempo:.6f} sec")
```

#Acceso a datos usando índices

Este bloque define una base de datos como una lista de listas, donde cada fila contiene el nombre, la edad y el animal favorito de una persona. Se crean dos diccionarios para identificar la posición de filas y columnas. Finalmente, se accede a un dato específico mostrando el animal favorito de Kevin.

```python
base_de_datos = [
["Oscar", 21, "Gatos"],
["Paula", 21, "Tiburones"],
["Jonatan", 24, "Hurones"],
["Kevin", 22, "Periquitos"],
["Lucero", 21, "Perros"],
["Israel", 21, "Lobos"],
["Sergio", 21, "Perros"],
["Vahir", 21, "Leones"],
["Diego", 21, "Pinguinos"],
]

columnas = {"Nombre":0, "Edad":1, "Animal":2}
filas    = {"Oscar":0, "Paula":1, "Jonatan":2, "Kevin":3, "Lucero":4, "Israel":5, "Sergio":6, "Vahir":7, "Diego":8}
print(base_de_datos[filas["Kevin"]][columnas["Animal"]])
```

#Manipulación de datos con Pandas

Este bloque transforma la base de datos en un DataFrame utilizando la librería Pandas. Se asignan nombres a las columnas y un índice numérico a cada fila. Posteriormente, se imprime la tabla completa para su visualización. Finalmente, se realiza un filtrado para obtener el animal favorito de Kevin.

```python
dF = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index=range(1,10))
print(dF)
print(dF.Animal[dF['Nombre'] == 'Kevin'])
```
