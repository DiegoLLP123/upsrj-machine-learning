# Importación de librerías

Vamos a importar las librerías necesarias para este ejercicio. Numpy nos permitirá manejar los datos como arreglos y realizar operaciones matemáticas fácilmente. Matplotlib nos ayudará a graficar los datos y visualizar su comportamiento. De sklearn usamos LinearRegression para crear el modelo de regresión lineal, y mean_squared_error y r2_score para evaluar qué tan bien funciona nuestro modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Creación de datos de prueba

Creamos un conjunto de datos de ejemplo para probar nuestro modelo. La variable x será nuestra entrada (independiente) y y será la salida (dependiente) que queremos predecir. Notemos que sklearn requiere que x sea un arreglo 2D, por eso usamos reshape(-1,1). Estos datos son simples para poder observar claramente la relación lineal.

```python
# y = m * x + b

# Variable independiente
# SciKit-learn requiere que se usen arreglos de 20
x = np.array((1, 2, 3, 4, 5)).reshape(-1, 1)
print(x)

# Variable dependiente
y = np.array((2, 4, 5, 4, 5))
print(y)
```

# Graficación de datos originales

Antes de entrenar el modelo, es buena idea visualizar los datos. Aquí hacemos un diagrama de dispersión para observar la relación entre x e y. Agregamos etiquetas, título y cuadrícula para que el gráfico sea más fácil de interpretar y entender si hay una tendencia lineal.

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Datos originales")
plt.grid(True)
plt.show()
```

# Generación y alimentación del modelo

Creamos el modelo de regresión lineal y lo entrenamos usando nuestros datos. El modelo aprende la relación entre x e y, y nos da los parámetros: la pendiente m, que indica cuánto cambia y por cada unidad de x, y el intercepto b, que indica dónde corta la recta al eje Y.

```python
# Generamos nuestro modelo de regresión linear
model = LinearRegression()

# Alimentamos nuestro modelo
model.fit(x, y)

# Parámetros: m (peso), b (sesgo)
m = model.coef_[0]
b = model.intercept_

print(f"Pendiente (m): {m : .2f}")
print(f"Intercepto (b): {b : .2f}")
```

# Predicción y evaluación del modelo

Con el modelo entrenado, predecimos los valores de y a partir de x. Evaluamos qué tan bien se ajusta el modelo usando dos métricas: el Mean Squared Error (MSE), que indica el error promedio al cuadrado entre los valores reales y predichos, y el R2 Score, que nos dice qué tan bien el modelo explica la variación de los datos, donde 1 significa un ajuste perfecto.

```python
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Meean Squared Error: {mse : .2f}")
print(f"R2 Score: {r2: .2f}")
```

# Graficación del modelo

Finalmente, graficamos los datos originales y la línea de regresión que aprendió el modelo. Esto nos permite ver visualmente cómo se ajusta la recta a los datos reales. Se añaden etiquetas, título, cuadrícula y leyenda para que el gráfico sea claro y fácil de interpretar.

```python
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred, color="red", label="Regresión lineal")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresión linear simple")
plt.grid(visible = True)
plt.show()
```
